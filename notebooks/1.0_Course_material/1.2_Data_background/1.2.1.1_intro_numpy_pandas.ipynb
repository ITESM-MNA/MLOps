# Dot Product

```python
def dot_product(a, b):
    result = 0
    for i in range(len(a)):
        result = result + a[i]*b[i]
    return result 
```

```python
np.random.random(5)
```

```python
a = np.random.randn(100)
b = np.random.randn(100)
print(a[:10])
```

```python
import time

start = time.time()

dot_product(a, b)


total = time.time() - start
print("total time: ", total)
```

```python
import time

start = time.time()

# dot product with numpy
np.dot(a, b)

total = time.time() - start
print("total time: ", total)
```
